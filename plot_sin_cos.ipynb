```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

class Man:
    """functions of class """

    
        
    def sin():
        # data
        x = np.arange(0, 6, 0.1)
        y = np.sin(x)
        plt.plot(x, y, label="sin")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('sin')
        plt.plot(x, y)
        plt.show()

    def cos():
        # datas
        x = np.arange(0, 6, 0.1)         
        y2 = np.cos(x)        
        plt.plot(x, y2, linestyle = "--", label="cos")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('cos')
        plt.legend()
        plt.show()
        
    def sincos():
        # datas
        x = np.arange(0, 6, 0.1)
        y1 = np.sin(x)
        y2 = np.cos(x)
        # plot graphs
        plt.plot(x, y1, label="sin")
        plt.plot(x, y2, linestyle = "--", label="cos")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('sin & cos')
        plt.legend()
        plt.show()

```

```python
Man.sin()
```

```python
Man.cos()
```

```python
Man.sincos()
```
