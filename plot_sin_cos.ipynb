```python
# coding: utf-8
import numpy as np
import matplotlib.pyplot as plt

class plot10530115:
    """functions of class """

        
    def __init__(self,x):
        self.x = np.arange(0, 6, 0.1)   

    def sin(self):
        
        y = np.sin(self.x)
        plt.plot(self.x, y, label="sin")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('sin')
        plt.plot(self.x, y)
        plt.show()

    def cos(self):
        # datas
               
        y2 = np.cos(self.x)        
        plt.plot(self.x, y2, linestyle = "--", label="cos")
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.title('cos')
        plt.legend()
        plt.show()
        
   

```

```python
m = plot10530115("")
m.sin()
m.cos()
```
